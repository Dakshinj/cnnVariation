```python
def CNN_layer(x_input, kernel_height, kernel_width, kernel_count, conv_name, stride = 1, padding = 'SAME'):

    channels = int(x_input.get_shape()[3])

    with tf.variable_scope(conv_name) as scope:
            
        W = tf.get_variable('W', shape = [kernel_height, kernel_width, channels, kernel_count],
            initializer = tf.contrib.layers.xavier_initializer(seed=0))
        B = tf.get_variable('B', shape = [kernel_count], initializer = tf.constant_initializer(0.0))
        # CNN layer
        conv = tf.nn.conv2d(x_input, W, strides = [1, stride, stride, 1], padding = padding)
        z = tf.nn.bias_add(conv, B)
        activation = tf.nn.relu(z)
        
        return activation
```

```python
def pool_layer(x_input, pool_name, kernel_height = 3, kernel_width = 3, stride = 2, padding = 'SAME'):

    return tf.nn.max_pool(x_input, ksize = [1, kernel_height, kernel_width, 1],
            strides = [1, stride, stride, 1], padding = padding, name = pool_name)
```

```python
def fully_connected_layer(x_input, fc_input, fc_output, fc_name, final_layer = False):

    with tf.variable_scope(fc_name) as scope:

        W = tf.get_variable('weights', shape = [fc_input, fc_output],
            initializer = tf.contrib.layers.xavier_initializer(seed=0))
        B = tf.get_variable('biases', shape = [fc_output], initializer = tf.constant_initializer(0.0))

        z = tf.nn.bias_add(tf.matmul(x_input, W), B)

        if final_layer:
            # No relu if it is the fine layer
            return z
        else:
            a = tf.nn.relu(z)
            return a
```

```python
def inception_layer(x_input, c1_kernel_count, reduce_c3_kernel_count, c3_kernel_count, reduce_c5_kernel_count, 
                    c5_kernel_count, c1_pool_kernel_count, inception_name):

    with tf.variable_scope(inception_name) as scope:

        c1 = CNN_layer(x_input, kernel_height = 1, kernel_width = 1,
                    kernel_count = c1_kernel_count, conv_name = '{}conv_1X1'.format(inception_name))

        
        # Reduction layer
        reduce_c3 = CNN_layer(x_input, kernel_height = 1, kernel_width = 1,
                    kernel_count = reduce_c3_kernel_count, conv_name = '{}reduce_conv_3X3'.format(inception_name))

        reduce_c5 = CNN_layer(x_input, kernel_height = 1, kernel_width = 1,
                    kernel_count = reduce_c5_kernel_count, conv_name = '{}reduce_conv_5X5'.format(inception_name))
        
        p1 = pool_layer(x_input, stride =1, pool_name = '{}_pooling'.format(inception_name))
        
        
        # CNN layer - 3X3 and 5X5 
        c3 = CNN_layer(reduce_c3, kernel_height = 3, kernel_width = 3,
                    kernel_count = c3_kernel_count, conv_name = '{}conv_3X3'.format(inception_name))
        
        c5 = CNN_layer(reduce_c5, kernel_height = 5, kernel_width = 5,
                    kernel_count = c5_kernel_count, conv_name = '{}conv_5X5'.format(inception_name))

        c1_pool = CNN_layer(p1, kernel_height = 1, kernel_width = 1,
                    kernel_count = c1_pool_kernel_count, conv_name = '{}_pool_conv'.format(inception_name))

        return tf.concat([c1, c3, c5, c1_pool], axis = 3, name='{}_merge'.format(inception_name))
```

```python
# tf_x - neural net input, tf_y - neural net output
tf.reset_default_graph() # resets the graph weigths
#print(train_x.shape[1])
#print(train_x.shape[2])
#print(train_x.shape[3])
#print(train_y.shape[1])
tf_x = tf.placeholder(tf.float32, shape=(None, train_x.shape[1], train_x.shape[2], train_x.shape[3]), name='tf_x')
tf_y = tf.placeholder(tf.float32, shape=(None, train_y.shape[1]), name='tf_y')
hold_prob = tf.placeholder(tf.float32)
print("Input Dimensions")
print(tf_x)
print("Output Dimensions")
print(tf_y)
```

## Neural Architecture

```python
# Start with Conv and pooling layer ---> c - conv, p - pooling, fc - fully connected, d - dropout
c1 = CNN_layer(tf_x, kernel_height = 5, kernel_width=5, kernel_count=64, conv_name='conv_layer_1')
p1 = pool_layer(c1, pool_name='pooling_layer_1')
```

```python
# Inception layers
# inception1 = inception_layer(p1, 64, 96, 128, 16, 32, 32, inception_name = 'inception1')
inception2 = inception_layer(p1, 128, 128, 192, 32, 96, 64, inception_name = 'inception2')
p2 = pool_layer(inception2, pool_name = 'pooling_layer_2')
```

```python
# hold_prob - dropout percentage
d1 = tf.nn.dropout(p2, keep_prob = hold_prob)
flatten = tf.contrib.layers.flatten(d1)

fc1 = fully_connected_layer(flatten, fc_input = int(flatten.get_shape()[-1]), fc_output = 8, 
               final_layer = True, fc_name = 'fc1')
```

```python
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=fc1, labels=tf_y))
print(cost)
optimizer = tf.train.AdamOptimizer(learning_rate=0.001).minimize(cost)
```

```python
def batch_generation(size, seed, x_data, y_data):

    example_count = x_data.shape[0]
    np.random.seed(seed)
    new_order = list(np.random.permutation(example_count))
    x_data = x_data[new_order,:,:,:]
    y_data = y_data[new_order,:]
    all_batch = []
    
    for i in range(0, math.floor(example_count/size)):
        xbatch = x_data[i * size : i * size + size,:,:,:]
        ybatch = y_data[i * size : i * size + size,:]
        all_batch.append((xbatch, ybatch))
    
    if example_count % size != 0:
        xbatch = x_data[math.floor(example_count/size) * size : example_count,:,:,:]
        ybatch = y_data[math.floor(example_count/size) * size : example_count,:]
        all_batch.append((xbatch, ybatch))
    
    return all_batch
```

```python
seed = 3
example_count = train_x.shape[0]
x_cor = train_x.shape[1]
y_cor = train_x.shape[2]
color_cor = train_x.shape[3]
y_count = train_y.shape[1]                            
total_cost = []
epochs = 20
batch_size = 32

init = tf.global_variables_initializer()
     
with tf.Session() as sess:        
    sess.run(init)        
    for epoch in range(epochs):
        batch_cost = 0.
        batch_count = int(example_count/batch_size)
        seed = seed + 1
        batches = batch_generation(batch_size, seed, train_x, train_y)
        for batch in batches:
            (x_data, y_data) = batch
            _ , model_cost = sess.run([optimizer, cost], feed_dict={tf_x: x_data, tf_y: y_data, hold_prob:1.0})
            batch_cost += model_cost / batch_count
            total_cost.append(batch_cost)
            
    # plot the cost
    plt.plot(np.squeeze(total_cost))
    plt.ylabel('Cost')
    plt.xlabel('Iterations')
    plt.show()
    
    # Calculate the correct predictions
    saver = tf.train.Saver()
    saver.save(sess, 'natural_image_model') 
    predict_op = tf.argmax(fc1,1)
    correct_prediction = tf.equal(predict_op, tf.argmax(tf_y,1))
    
    # Calculate accuracy on the test set
    accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))
    print("Accuracy:")
    # train_accuracy = accuracy.eval({tf_x: train_x, tf_y: train_y, hold_prob:1.0})
    # print("Train Accuracy:", train_accuracy)
    test_accuracy = accuracy.eval({tf_x: test_x, tf_y: test_y, hold_prob:1.0})
    print("Test Accuracy:", test_accuracy)
```
