```python
tf.reset_default_graph()
# print(train_x.shape[1])
# print(train_x.shape[2])
# print(train_x.shape[3])
# print(train_y.shape[1])
tf_x = tf.placeholder(tf.float32, shape=(None, train_x.shape[1], train_x.shape[2], train_x.shape[3]), name='tf_x')
tf_y = tf.placeholder(tf.float32, shape=(None, train_y.shape[1]), name='tf_y')
print("Input Dimensions")
print(tf_x)
print("Output Dimensions")
print(tf_y)

# weights
weight1 = tf.get_variable('weight1', shape=(8,8,3,64), initializer=tf.contrib.layers.xavier_initializer(seed=0))
weight2 = tf.get_variable('weight2', shape=(6,6,64,32), initializer=tf.contrib.layers.xavier_initializer(seed=0))
weight3 = tf.get_variable('weight3', shape=(4,4,32,8), initializer=tf.contrib.layers.xavier_initializer(seed=0))
weight4 = tf.get_variable('weight4', shape=(2,2,8,12), initializer=tf.contrib.layers.xavier_initializer(seed=0))
```

```python
# Block - 1
Z1 = tf.nn.conv2d(tf_x, weight1, strides=[1, 1, 1, 1], padding='SAME')
A1 = tf.nn.relu(Z1)
P1 = tf.nn.max_pool(A1, ksize=[1, 4, 4, 1], strides=[1, 4, 4, 1], padding='SAME')
# Block - 2
Z2 = tf.nn.conv2d(P1, weight2, strides=[1, 1, 1, 1], padding='SAME')
A2 = tf.nn.relu(Z2)
P2 = tf.nn.max_pool(A2, ksize=[1, 4, 4, 1], strides=[1, 4, 4, 1], padding='SAME')
# Block - 3
Z3 = tf.nn.conv2d(P2, weight3, strides=[1, 1, 1, 1], padding='SAME')
A3 = tf.nn.relu(Z3)
P3 = tf.nn.max_pool(A3, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')
# Block - 4
Z4 = tf.nn.conv2d(P3, weight4, strides=[1, 1, 1, 1], padding='SAME')
A4 = tf.nn.relu(Z4)
P4 = tf.nn.max_pool(A4, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')

# Fully Connected Block
P4 = tf.contrib.layers.flatten(P4)
# 8 classes
Z5 = tf.contrib.layers.fully_connected(P4, 8, activation_fn=None)
```

```python
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=Z5, labels=tf_y))
optimizer = tf.train.AdamOptimizer(learning_rate=0.001).minimize(cost)
```

```python
# Reference
# https://www.deeplearning.ai/deep-learning-specialization/ 
def batch_generation(size, seed, x_data, y_data):

    example_count = x_data.shape[0]
    np.random.seed(seed)
    new_order = list(np.random.permutation(example_count))
    x_data = x_data[new_order,:,:,:]
    y_data = y_data[new_order,:]
    all_batch = []
    
    for i in range(0, math.floor(example_count/size)):
        xbatch = x_data[i * size : i * size + size,:,:,:]
        ybatch = y_data[i * size : i * size + size,:]
        all_batch.append((xbatch, ybatch))
    
    if example_count % size != 0:
        xbatch = x_data[math.floor(example_count/size) * size : example_count,:,:,:]
        ybatch = y_data[math.floor(example_count/size) * size : example_count,:]
        all_batch.append((xbatch, ybatch))
    
    return all_batch
```

```python
seed = 3
example_count = train_x.shape[0]
x_cor = train_x.shape[1]
y_cor = train_x.shape[2]
color_cor = train_x.shape[3]
y_count = train_y.shape[1]                            
total_cost = []
epochs = 40
batch_size = 128

init = tf.global_variables_initializer()
     
with tf.Session() as sess:        
    sess.run(init)        
    for epoch in range(epochs):
        batch_cost = 0.
        batch_count = int(example_count/batch_size)
        seed = seed + 1
        batches = batch_generation(batch_size, seed, train_x, train_y)
        for batch in batches:
            (x_data, y_data) = batch
            _ , model_cost = sess.run([optimizer, cost], feed_dict={tf_x: x_data, tf_y: y_data})
            batch_cost += model_cost / batch_count
            total_cost.append(batch_cost)
            
    # plot the cost
    plt.plot(np.squeeze(total_cost))
    plt.ylabel('Cost')
    plt.xlabel('Iterations')
    plt.show()
    
    # Calculate the correct predictions
    saver = tf.train.Saver()
    saver.save(sess, 'natural_image_model') 
    predict_op = tf.argmax(Z5,1)
    correct_prediction = tf.equal(predict_op, tf.argmax(tf_y,1))
    
    # Calculate accuracy on the test set
    accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))
    print("Accuracy:")
    train_accuracy = accuracy.eval({tf_x: train_x, tf_y: train_y})
    print("Train Accuracy:", train_accuracy)
    test_accuracy = accuracy.eval({tf_x: test_x, tf_y: test_y})
    print("Test Accuracy:", test_accuracy)
```
